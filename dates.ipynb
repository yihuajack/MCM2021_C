```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
```

```python
all_data = pd.read_csv('./data/2021MCMProblemC_DataSet.csv', encoding='utf-8', index_col='Detection Date')
# all_data
## for the first question, only keep track of the sightings in 2019 and 2020
all_data.index = pd.to_datetime(all_data.index, errors='coerce')
all_data = all_data.sort_values(by = 'Detection Date', ascending=True)
data = all_data[all_data.index > '2000/1/1']
data
```

```python
y = {}
for d in data.index:
    if not d.year in y:
        y[d.year] = 1
    else:
        y[d.year] += 1
print(y)

m = {}
for d in data.index:
    if not d.month in m:
        m[d.month] = 1
    else:
        m[d.month] += 1
print(m)
```

```python

#facecolor:è¡¨é¢çš„é¢œè‰²;edgecolor:è¾¹æ¡†çš„é¢œè‰²
plt.bar(m.keys(), m.values(), facecolor = '#9999ff', edgecolor = 'white')
plt.xlabel('month')
plt.ylabel('counts')
plt.title('number of reported sighting for each month')

plt.show()
```
